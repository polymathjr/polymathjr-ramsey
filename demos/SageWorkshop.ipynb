```python
def search_ramsey(s, t):
    n = 2
    while True:
        print(f'searching graphs of size {n}...')

        found_LB = False
        
        for g in graphs(n):
            
            if g.clique_number() < s and g.complement().clique_number() < t:
                print(f'R({s}, {t}) > {n}:')
#                 g.show()
                n += 1
                found_LB = True
                break
        
        if not found_LB:
            return n
    
search_ramsey(3, 4)
```

## Solving Integer Programs

```python
p = MixedIntegerLinearProgram(maximization=False, solver = "GLPK")
w = p.new_variable(integer=True, nonnegative=True)

p.add_constraint(w[0] + w[1] + w[2] - 14*w[3] == 0)
p.add_constraint(w[1] + 2*w[2] - 8*w[3] == 0)
p.add_constraint(2*w[2] - 3*w[3] == 0)
p.add_constraint(w[0] - w[1] - w[2] >= 0)
p.add_constraint(w[3] >= 1)

p.set_objective(w[3])
p.show()
```

```python
p.solve()
```

```python
for i, v in sorted(p.get_values(w, convert=ZZ, tolerance=1e-3).items()):
    print(f'w_{i} = {v}')
```
